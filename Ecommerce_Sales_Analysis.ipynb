```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/Sample data.csv', encoding='latin1')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

# Converting date columns

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year']=data['Order Date'].dt.year
data['Order Day of Week']=data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# Montly sales analysis

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=sales_by_month, x='Order Month', y='Sales')
plt.title('Monthly Sales Analysis')
plt.show()
```

# Sales by Category

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
plt.figure(figsize=(8, 8))
colors = sns.color_palette('pastel')[0:len(sales_by_category)]
plt.pie(sales_by_category['Sales'], autopct=lambda pct: '{:.1f}%\n{}'.format(pct, sales_by_category['Category'][min(range(len(sales_by_category['Sales'])), key=lambda i: abs(sales_by_category['Sales'][i]/sales_by_category['Sales'].sum()*100 - pct))]), startangle=140, colors=colors)
plt.title('Sales by Category')
plt.show()
```

# Sales analysis by Sub Category

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_subcategory
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_subcategory, x='Sub-Category', y='Sales')
plt.title('Sales by Sub-Category')
plt.xticks(rotation=90)
plt.show()
```

# Montly profit analysis

```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=profit_by_month, x='Order Month', y='Profit')
plt.title('Monthly Profit Analysis')
plt.show()
```

# Profit by Category


```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
plt.figure(figsize=(8, 8))
colors = sns.color_palette('pastel')[0:len(profit_by_category)]
plt.pie(profit_by_category['Profit'], autopct=lambda pct: '{:.1f}%\n{}'.format(pct, profit_by_category['Category'][min(range(len(profit_by_category['Profit'])), key=lambda i: abs(profit_by_category['Profit'][i]/profit_by_category['Profit'].sum()*100 - pct))]), startangle=140, colors=colors)
plt.title('Profit by Category')
plt.show()
```

# Profit by Sub Category

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_subcategory
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=profit_by_subcategory, x='Sub-Category', y='Profit')
plt.title('Profit by Sub-Category')
plt.xticks(rotation=90)
plt.show()
```

Sales and Profit - Customer Segment

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment
```

```python
sales_profit_by_segment_melted = sales_profit_by_segment.melt(id_vars='Segment', var_name='Metric', value_name='Amount')

plt.figure(figsize=(10, 6))
sns.barplot(data=sales_profit_by_segment_melted, x='Segment', y='Amount', hue='Metric', palette='viridis')
plt.title('Sales and Profit by Customer Segment')
plt.ylabel('Amount')
plt.show()
```

# Sales to profit ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales to Profit Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
display(sales_profit_by_segment)
```
